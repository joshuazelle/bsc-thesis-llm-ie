### look for missing websites

```python
import requests
import pandas as pd


def find_website(company_name, api_key, cse_id):
    # Refine the query to include 'semiconductor', 'website', and 'English'
    query = f"{company_name} semiconductor website English"
    url = f"https://www.googleapis.com/customsearch/v1?key={api_key}&cx={cse_id}&q={query}"
    response = requests.get(url)
    result = response.json()

    if "items" in result:
        for item in result["items"]:
            link = item["link"]
            # Exclude certain domains like LinkedIn and Wikipedia
            if "linkedin.com" not in link and "wikipedia.org" not in link:
                return link
    return None


for index, row in missing.iterrows():
    website_address = find_website(row["name_internat"], api_key, cse_id)
    if website_address:
        # Update the original DataFrame
        missing.at[index, "website_address"] = website_address
```
