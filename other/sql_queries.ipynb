```sql

CREATE TEMPORARY VIEW gt_ascii_map AS
  SELECT *
  FROM '/home/zelle/development/projects/ascii/my_dev/georgetown_orbis/map_ascii_orbis_gt.csv';

```sql
CREATE temporary VIEW websites AS
· SELECT 
·     gt_ascii_map.provider_id,
      gt_ascii_map.ascii_id_company,
      ascii_ref_clean.orbis_company_contact_info.name_internat, 
·     ascii_ref_clean.orbis_company_contact_info.website_address
· FROM 
·     gt_ascii_map
· JOIN 
·     ascii_ref_clean.orbis_company_contact_info
· ON 
·     gt_ascii_map.ascii_id_company = ascii_ref_clean.orbis_company_contact_info.ascii_id_company;

## for exploring pipeline output

```sql
select count(*) from ascii_commoncrawl.nodes_meta as meta where meta.seed_nodes='Georgetown_v1.0.0';

```sql
select count(*) from (select * from ascii_commoncrawl.nodes_meta as meta where meta.seed_nodes='Georgetown_v1.0.0')
  where url_surtkey not like '%google%' limit 10;

```sql
SELECT COUNT(DISTINCT url_host_registered_domain) AS distinct_domain_count
  FROM (
    SELECT *
    FROM ascii_commoncrawl.nodes_meta AS meta
    WHERE meta.seed_nodes = 'Georgetown_v1.0.0'
  ) AS filtered_meta
  WHERE url_surtkey NOT LIKE '%google%';
┌───────────────────────┐
│ distinct_domain_count │
│         int64         │
├───────────────────────┤
│                   166 │

```sql
SELECT AVG(domain_count) AS average_per_domain
  FROM (
    SELECT url_host_registered_domain, COUNT(*) AS domain_count
    FROM (
      SELECT *
      FROM ascii_commoncrawl.nodes_meta AS meta
      WHERE meta.seed_nodes = 'Georgetown_v1.0.0'
    ) AS filtered_meta
    WHERE url_surtkey NOT LIKE '%google%'
    GROUP BY url_host_registered_domain
  ) AS counts;
┌────────────────────┐
│ average_per_domain │
│       double       │
├────────────────────┤
│  6170.506024096386 │

```sql
pragma table_info(ascii_commoncrawl.edges);
┌───────┬──────────────────────┬─────────────────────────────────────────────────────────────────────────┬─────────┬────────────┬─────────┐
│  cid  │         name         │                                  type                                   │ notnull │ dflt_value │   pk    │
│ int32 │       varchar        │                                 varchar                                 │ boolean │  varchar   │ boolean │
├───────┼──────────────────────┼─────────────────────────────────────────────────────────────────────────┼─────────┼────────────┼─────────┤
│     0 │ src_url_surt         │ VARCHAR                                                                 │ false   │            │ false   │
│     1 │ src_url              │ VARCHAR                                                                 │ false   │            │ false   │
│     2 │ src_url_surt_host    │ VARCHAR                                                                 │ false   │            │ false   │
│     3 │ src_url_redirected…  │ VARCHAR                                                                 │ false   │            │ false   │
│     4 │ dst_urls             │ STRUCT(url VARCHAR, surt_host VARCHAR, attributes MAP(VARCHAR, VARCHA…  │ false   │            │ false   │
│     5 │ content              │ VARCHAR                                                                 │ false   │            │ false   │
│     6 │ crawl_id             │ VARCHAR                                                                 │ false   │            │ false   │
│     7 │ seed_nodes           │ VARCHAR                                                                 │ false   │            │ false   │
└───────┴────────────────

```sql
D pragma table_info(ascii_commoncrawl.nodes_meta);
┌───────┬────────────────────────────┬─────────┬─────────┬────────────┬─────────┐
│  cid  │            name            │  type   │ notnull │ dflt_value │   pk    │
│ int32 │          varchar           │ varchar │ boolean │  varchar   │ boolean │
├───────┼────────────────────────────┼─────────┼─────────┼────────────┼─────────┤
│     0 │ url_host_registered_domain │ VARCHAR │ false   │            │ false   │
│     1 │ url_surtkey                │ VARCHAR │ false   │            │ false   │
│     2 │ url                        │ VARCHAR │ false   │            │ false   │
│     3 │ warc_filename              │ VARCHAR │ false   │            │ false   │
│     4 │ warc_record_offset         │ INTEGER │ false   │            │ false   │
│     5 │ warc_record_length         │ INTEGER │ false   │            │ false   │
│     6 │ crawl_id                   │ VARCHAR │ false   │            │ false   │
│     7 │ seed_nodes                 │ VARCHAR │ false   │            │ false   

pragma table_info(ascii_commoncrawl.seed_nodes);
┌───────┬────────────────────┬─────────┬─────────┬────────────┬─────────┐
│  cid  │        name        │  type   │ notnull │ dflt_value │   pk    │
│ int32 │      varchar       │ varchar │ boolean │  varchar   │ boolean │
├───────┼────────────────────┼─────────┼─────────┼────────────┼─────────┤
│     0 │ ascii_id_company   │ VARCHAR │ false   │            │ false   │
│     1 │ seed_node_url      │ VARCHAR │ false   │            │ false   │
│     2 │ seed_node_url_surt │ VARCHAR │ false   │            │ false   │
│     3 │ seeds              │ VARCHAR │ false   │            │ false   │
└───────┴────────────────────┴─────────┴─────────┴────────────┴─────────┘

### filter urls

```sql
SELECT count(*)
            FROM ascii_commoncrawl.nodes_meta AS meta
            WHERE meta.seed_nodes = 'Georgetown_v1.0.0' AND meta.crawl_id='CC-MAIN-2021-49';
┌──────────────┐
│ count_star() │
│    int64     │
├──────────────┤
│      2143908 │

select count(*) from ascii_commoncrawl.edges where crawl_id = 'CC-MAIN-2021-49' and seed_nodes = 'Georgetown_v1.0.0';
┌──────────────┐
│ count_star() │
│    int64     │
├──────────────┤
│      2143908 │
└──────────────┘
select count(*) from ascii_commoncrawl.edges where crawl_id = 'CC-MAIN-2021-49' and seed_nodes = 'Georgetown_v1.0.0' and src_url not like '%google%';
┌──────────────┐
│ count_star() │
│    int64     │
├──────────────┤
│       803746 │
└──────────────┘
SELECT count(*)
              FROM ascii_commoncrawl.nodes_meta AS meta
              WHERE meta.seed_nodes = 'Georgetown_v1.0.0' AND meta.crawl_id='CC-MAIN-2021-49'and url not like '%google%';
┌──────────────┐
│ count_star() │
│    int64     │
├──────────────┤
│       803746 │
└──────────────┘

select count(*) from ascii_commoncrawl.edges where crawl_id = 'CC-MAIN-2021-49' and seed_nodes = 'Georgetown_v1.0.0' and src_url_surt_host not like '%google%';
┌──────────────┐
│ count_star() │
│    int64     │
├──────────────┤
│       804092 │
└──────────────┘


ok so thats the same thats good, and how do i best go ahead? join them? yeah lets try and then do the filtering

D pragma table_info(ascii_commoncrawl.edges);
┌───────┬───────────────────────┬────────────────────────────────────────────────────────────────────────────┬─────────┬────────────┬─────────┐
│  cid  │         name          │                                    type                                    │ notnull │ dflt_value │   pk    │
│ int32 │        varchar        │                                  varchar                                   │ boolean │  varchar   │ boolean │
├───────┼───────────────────────┼────────────────────────────────────────────────────────────────────────────┼─────────┼────────────┼─────────┤
│     0 │ src_url_surt          │ VARCHAR                                                                    │ false   │            │ false   │
│     1 │ src_url               │ VARCHAR                                                                    │ false   │            │ false   │
│     2 │ src_url_surt_host     │ VARCHAR                                                                    │ false   │            │ false   │
│     3 │ src_url_redirected_to │ VARCHAR                                                                    │ false   │            │ false   │
│     4 │ dst_urls              │ STRUCT(url VARCHAR, surt_host VARCHAR, attributes MAP(VARCHAR, VARCHAR))[] │ false   │            │ false   │
│     5 │ content               │ VARCHAR                                                                    │ false   │            │ false   │
│     6 │ crawl_id              │ VARCHAR                                                                    │ false   │            │ false   │
│     7 │ seed_nodes            │ VARCHAR                                                                    │ false   │            │ false   │
└───────┴───────────────────────┴────────────────────────────────────────────────────────────────────────────┴─────────┴────────────┴─────────┘
D pragma table_info(ascii_commoncrawl.nodes_meta);
┌───────┬────────────────────────────┬─────────┬─────────┬────────────┬─────────┐
│  cid  │            name            │  type   │ notnull │ dflt_value │   pk    │
│ int32 │          varchar           │ varchar │ boolean │  varchar   │ boolean │
├───────┼────────────────────────────┼─────────┼─────────┼────────────┼─────────┤
│     0 │ url_host_registered_domain │ VARCHAR │ false   │            │ false   │
│     1 │ url_surtkey                │ VARCHAR │ false   │            │ false   │
│     2 │ url                        │ VARCHAR │ false   │            │ false   │
│     3 │ warc_filename              │ VARCHAR │ false   │            │ false   │
│     4 │ warc_record_offset         │ INTEGER │ false   │            │ false   │
│     5 │ warc_record_length         │ INTEGER │ false   │            │ false   │
│     6 │ crawl_id                   │ VARCHAR │ false   │            │ false   │
│     7 │ seed_nodes                 │ VARCHAR │ false   │            │ false   │
└───────┴────────────────────────────┴─────────┴─────────┴────────────┴─────────┘

check duplicates: 

```sql
SELECT url, COUNT(*)
  FROM ascii_commoncrawl.nodes_meta
  WHERE crawl_id = 'CC-MAIN-2021-49'
  AND seed_nodes = 'Georgetown_v1.0.0'
  GROUP BY url
  HAVING COUNT(*) > 1;

  29455 rows (40 shown)

How many unique companies were found and have urls?

```sql
SELECT count(distinct sn.ascii_id_company)
  FROM ascii_commoncrawl.edges e
  JOIN ascii_commoncrawl.seed_nodes sn ON e.src_url_surt_host = sn.seed_node_url_surt
  WHERE e.crawl_id = 'CC-MAIN-2021-49'
  AND e.seed_nodes = 'Georgetown_v1.0.0'
  AND e.src_url_surt_host NOT LIKE '%google%'
  AND sn.seeds = 'Georgetown_v1.0.0';
┌─────────────────────────────────────┐
│ count(DISTINCT sn.ascii_id_company) │
│                int64                │
├─────────────────────────────────────┤
│                                 168 │
└─────────────────────────────────────┘

whats the average urls per company
SELECT AVG(src_url_count) AS avg_src_urls_per_id
  FROM (
    SELECT sn.ascii_id_company, COUNT(DISTINCT e.src_url) AS src_url_count
    FROM ascii_commoncrawl.edges e
    JOIN ascii_commoncrawl.seed_nodes sn ON e.src_url_surt_host = sn.seed_node_url_surt
    WHERE e.crawl_id = 'CC-MAIN-2021-49'
    AND e.seed_nodes = 'Georgetown_v1.0.0'
    AND e.src_url_surt_host NOT LIKE '%google%'
    AND sn.seeds = 'Georgetown_v1.0.0'
    GROUP BY sn.ascii_id_company
  ) AS counts;
┌─────────────────────┐
│ avg_src_urls_per_id │
│       double        │
├─────────────────────┤
│  4590.4702380952385 │
└─────────────────────┘

SELECT MIN(src_url_count) AS min_urls_per_id, MAX(src_url_count) AS max_urls_per_id, MEDIAN(src_url_count) AS median_urls_per_id
  FROM (
    SELECT sn.ascii_id_company, COUNT(DISTINCT e.src_url) AS src_url_count
    FROM ascii_commoncrawl.edges e
    JOIN ascii_commoncrawl.seed_nodes sn ON e.src_url_surt_host = sn.seed_node_url_surt
    WHERE e.crawl_id = 'CC-MAIN-2021-49'
    AND e.seed_nodes = 'Georgetown_v1.0.0'
    AND e.src_url_surt_host NOT LIKE '%google%'
    AND sn.seeds = 'Georgetown_v1.0.0'
    GROUP BY sn.ascii_id_company
  ) AS counts;
┌─────────────────┬─────────────────┬────────────────────┐
│ min_urls_per_id │ max_urls_per_id │ median_urls_per_id │
│      int64      │      int64      │       double       │
├─────────────────┼─────────────────┼────────────────────┤
│               1 │          138852 │               66.0 │
└─────────────────┴─────────────────┴────────────────────┘
